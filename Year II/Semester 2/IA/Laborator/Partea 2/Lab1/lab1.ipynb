```python
list_images = []
for i in range(0, 9):
    img = np.load('images\car_%d.npy' % i)
    list_images.append(img)
    
images = np.array(list_images)
```

```python
images.shape
```

```python
sum_mat = images.sum() # np.sum(images)
```

```python
print(sum_mat)
```

```python
# Calculați suma valorilor pixelilor pentru fiecare imagine în parte
sum_per_image = np.sum(images, axis=(1, 2))
```

```python
sum_per_image
```

```python
# Afișați indexul imaginii cu suma maximă
sum_per_image.argmax()
```

```python
# Calculați imaginea medie și afișati-o
mean_image = np.mean(images, axis=0)
```

```python
mean_image.shape
```

```python
plt.imshow(mean_image, cmap='gray')
plt.show()
```

```python
plt.imshow(images[-2], cmap='gray')
plt.show()
```

```python
# Cu ajutorul funcției np.std(images_array), calculați deviația standard a imaginilor
std_image = np.std(images, axis=0)
```

```python
# (se scade imaginea medie și se împarte rezultatul la deviația standard
```

```python
# 9x400x600 mean = 400x600 std=400x600
```

```python
normalized_images = (images - mean_image) / std_image
```

```python
normalized_images.shape
```

```python
# Decupați fiecare imagine, afișând numai liniile cuprinse între 200 și 300, respectiv coloanele cuprinse între 280 și 400.
cropped_images = images[:, 200:300, 280:400]
for cropped_image in cropped_images:
    plt.imshow(cropped_image, cmap='gray')
    plt.show()
```
